```python
## Criar um alerta de erro usando a função alerta()

def alerta(nivel, base, etapa):
    now = str(datetime.now())
    msg = f"Falha no carregamento da base {base} na etapa {etapa}.\n{now}"

    if nivel == 1:
        title = 'ATENÇÃO: Alerta Baixo'
    elif nivel == 2:
        title = 'ATENÇÃO: Alerta Médio'
    elif nivel == 3:
        title = 'ATENÇÃO: Alerta Alto'
    else:
        print("Nivel",nivel,"não disponível!")

    notification.notify(
        title=title,
        message=msg,
        app_name='alerta',
        timeout=10
    )
```

```python
## Criar função para acessar API

def acessarAPI(url=str):
    response = requests.get(url)
    if response.status_code == 200:
        data_json = response.json()
        return response
    else:
        alerta(nivel=3, base="REST Countries", etapa="EXTRACAO")

```

```python
## Realizar conexão com a API

response = acessarAPI("https://restcountries.com/v3.1/all")
data = response.json()
df = pd.DataFrame(data)
df.head()
```

```python
## Extrair primeira base - Países

countries = pd.DataFrame([{'name':df['name']['common'],'continent':df['continents']} for df in data])
countries
```

```python
## Extrair segunda base - Idiomas

languages = pd.DataFrame(df['languages'])
languages
```

```python
## Extrair terceira base - Moedas

currencies = pd.DataFrame(df['currencies'])
currencies
```

# Tratamento das Bases
### Realizar o tratamento das bases coletadas via API

Resumo das alterações:
- Ajuste em nome de coluna
- Unstack
- Ajuste de missing
- Tratamento de colunas string
- Criação de novas tabelas


```python
# Base 1: countries - Tratamento da coluna continent

countries['continent'] = countries['continent'].str[0].str.strip('[]')
countries
```

```python
# Base 2: languages - unstack e retirada de missing values


languages = pd.DataFrame(languages['languages'].apply(lambda x: list(x.values()) if isinstance(x, dict) else None)).explode('languages')
languages
```

```python
# Base 3: currencies - unstack e retirada de missing values


currencies = pd.DataFrame(currencies['currencies']).map(lambda x:list(x.keys()) if isinstance(x,dict) else None).explode('currencies')
currencies
```

```python
# Base Original: df - inclusão de unidade no nome da coluna área e criação de subset tratado

df.rename(columns={'area':'area(km2)'},inplace=True)
df1=pd.DataFrame(countries)
df1['population']=df['population']
df1['area(km2)']=df['area(km2)'].round(1)

```

# Análises
### Extrair insights com base nas bases apresentadas e dados disponíveis na API

Resumo das tabelas criadas:
- Primeira base: Países - Ranking de densidade demográfica por país e continente
- Segunda base: Idiomas - Ranking de idiomas mais falados por país e por população
- Terceira base: Moedas - Ranking de moedas mais utilizadas por país e por território




```python
# Tabela 1: Ranking de densidade demográfica por país

df1['density'] = (df['population']/df['area(km2)']).round(1)
ranking1 = df1.sort_values(by='density', ascending=False)
ranking1
```

```python
# Tabela 2: Ranking de densidade demográfica por continente

ranking2 = ranking1.groupby(by="continent").sum().drop(columns=['name'])
ranking2['density']=ranking2['population']/ranking2['area(km2)']
ranking2=ranking2.sort_values(by='density', ascending=False)
ranking2
```

```python
# Tabela 3: Ranking de idiomas falados em mais países

ranking3 = pd.DataFrame(languages.value_counts()).rename(columns={'count':'countries'})
ranking3
```

```python
# Tabela 4: Ranking de idiomas falados por mais pessoas

ranking4 = df1.merge(languages, how='left', left_index=True, right_index=True).groupby(by='languages').sum().drop(columns=['density','area(km2)','continent','name']).sort_values(by='population', ascending=False)
ranking4
```

```python
# Tabela 5: Ranking de moedas usadas em mais países

ranking5 = pd.DataFrame(currencies.value_counts()).rename(columns={'count':'countries'})
ranking5
```

```python
# Tabela 5: Ranking de moedas usadas por território

ranking6 = df1.merge(currencies, how='left', left_index=True, right_index=True).groupby(by='currencies').sum().drop(columns=['density','population','continent','name']).sort_values(by='area(km2)', ascending=False)
ranking6
```

# Banco de Dados

### Criar funções para criar, salvar tabelas e acessar o banco de dados do projeto

```python
def tabelas_bd():
    '''
        Retorna um dataframe com as tabelas do banco de dados.
    '''
    conn = sqlite3.connect('coderhouse.db')

    # Executar uma consulta que retorna as informações do esquema do banco de dados
    query = "SELECT name FROM sqlite_master WHERE type='table'"
    schema = pd.read_sql_query(query, conn)

    conn.close()

    return schema
def salva_bd(df, nome_tabela):
    '''
        Salva dataframe df na tabela nome_tabela.
    '''
    conn = sqlite3.connect('coderhouse.db')

    # Escrever o DataFrame na tabela 'nome_tabela'
    df.to_sql(nome_tabela, conn, if_exists='replace', index=False)

    conn.close()

    return True
def carrega_bd(nome_tabela):
    '''
        Carrega tabela nome_tabela num dataframe. 
    '''
    conn = sqlite3.connect('coderhouse.db')

    # Executar uma consulta SELECT na tabela 'produtos' e converter em um DataFrame
    query = f"SELECT * FROM {nome_tabela}"
    df = pd.read_sql(query, conn)

    conn.close()

    return df
```

```python
# Salvar tabelas criadas no banco de dados

salva_bd(ranking1,'Ranking - Densidade por país')
salva_bd(ranking2,'Ranking - Densidade por continente')
salva_bd(ranking3,'Ranking - Idioma por país')
salva_bd(ranking4,'Ranking - Idioma por população')
salva_bd(ranking5,'Ranking - Moeda por país')
salva_bd(ranking6,'Ranking - Moeda por território')
salva_bd(currencies,'Lista de Moedas')
salva_bd(languages,'Lista de Idiomas')
salva_bd(countries,'Lista de Países')
salva_bd(df1,'Dataset Tratado')
```
